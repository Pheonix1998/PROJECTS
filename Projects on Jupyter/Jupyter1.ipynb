```python
# Personal Introduction Program (Smart Interactive Version)
import datetime  # Used to get the current time

# Ask for user's name
name = input("Enter your name: ")

# Ask for user's age
age = input("Enter your age: ")

# Ask for multiple hobbies
hobbies = input("Enter your hobbies (separated by commas): ")

# Convert the hobby input into a clean list
hobby_list = [h.strip().lower() for h in hobbies.split(",")]

# Get current hour to decide greeting
hour = datetime.datetime.now().hour

if hour < 12:
    greeting = "Good morning"
elif hour < 18:
    greeting = "Good afternoon"
else:
    greeting = "Good evening"

# Display basic information
print(f"\n{greeting}, {name}!")
print(f"You are {age} years old and your hobbies are: {', '.join(hobby_list)}.")

# Follow-up questions based on hobbies
if "music" in hobby_list:
    music_type = input("That's awesome! What kind of music do you like? ")
    print(f"Nice! {music_type.capitalize()} music is great! ðŸŽ¶")

if "sports" in hobby_list:
    sport_name = input("Cool! Which sport do you enjoy the most? ")
    print(f"{sport_name.capitalize()} is such an exciting sport! ðŸ…")

if "reading" in hobby_list:
    book_genre = input("Wonderful! What type of books do you enjoy reading? ")
    print(f"{book_genre.capitalize()} books can be really inspiring! ðŸ“š")

if "traveling" in hobby_list or "travel" in hobby_list:
    place = input("Traveling sounds fun! Whatâ€™s your favorite place youâ€™ve visited? ")
    print(f"Wow, {place} must be amazing to visit! ðŸŒ")

# End message
print("\nIt was great talking to you, " + name + "! Have a wonderful day! ðŸ˜Š")
```
