```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"E:\Developer's arena Internship\week 3\Video game sales.csv", encoding='latin1')

# Calculate total global sales
total_sales = df["Global_Sales"].sum()

# Print result
print("Total Global Sales (in millions):", total_sales)
```

```python
df.head(5)
```

```python
# Find the best-selling product
best_selling = df.loc[df["Global_Sales"].idxmax()]

# Print the result
print("Best-Selling Product:")
print("Name:", best_selling["Name"])
print("Platform:", best_selling["Platform"])
print("Genre:", best_selling["Genre"])
print("Publisher:", best_selling["Publisher"])
print("Global Sales (millions):", best_selling["Global_Sales"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ================================
# (a) Number of Games Released per Year
# ================================

# Clean Year column (remove missing values)
games_per_year = df['Year'].dropna().astype(int).value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.plot(games_per_year.index, games_per_year.values)
plt.title("Number of Games Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Games")
plt.grid(True)
plt.tight_layout()
plt.show()


# ================================
# (b) Global Sales by Genre
# ================================

sales_by_genre = df.groupby('Genre')['Global_Sales'].sum().sort_values()

plt.figure(figsize=(8,5))
plt.bar(sales_by_genre.index, sales_by_genre.values)
plt.title("Global Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Total Global Sales (Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
